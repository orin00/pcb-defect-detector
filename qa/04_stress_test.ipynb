```python
import threading
import time
import requests
import os

# 테스트 설정
BASE_URL = "http://localhost:8000/api"
img_sample = r"..\VOC_PCB\test_sample.jpg"
thread_count = 10
results_log = []
```

```python
def upload_task(user_id):
    start = time.time()
    try:
        with open(img_sample, 'rb') as f:
            files = {'file': f}
            response = requests.post(f"{BASE_URL}/detect/", files=files)
            duration = time.time() - start
            log = f"User {user_id}: Status {response.status_code} ({duration:.2f}s)"
    except Exception as e:
        log = f"User {user_id} 에러: {e}"
    
    print(log)
    results_log.append(log)
```

```python
# 1. 테스트 실행
print(f"=== [QA] {thread_count}명 동시 업로드 테스트 시작 ===")
threads = []
for i in range(thread_count):
    t = threading.Thread(target=upload_task, args=(i,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
print("=== 동시성 테스트 완료 ===")

# 2. 결과 저장 로직
save_path = r"..\qa\model_qa_results"
if not os.path.exists(save_path):
    os.makedirs(save_path)

file_name = os.path.join(save_path, "concurrency_test_report.txt")

with open(file_name, 'w', encoding='utf-8') as f:
    f.write(f"=== 동시성 테스트 리포트 ({thread_count} Threads) ===\n")
    f.write(f"테스트 시간: {time.strftime('%Y-%m-%d %H:%M:%S')}\n")
    f.write("-" * 50 + "\n")
    for line in results_log:
        f.write(line + "\n")
    f.write("-" * 50 + "\n")
    f.write("=== 테스트 완료 ===\n")

print(f"정밀 성능 평가 결과가 저장되었습니다: {file_name}")
```
